```python
import time

def gcd(m, n):
  m_val = min(m, n);
  n_val = max(m, n);

  while m_val > 0:
    t = m_val;
    m_val = n_val % m_val;
    n_val = t;

  return n_val;
m = 12000000
n = 76000000


start_time = time.time()
result = gcd(m, n)
end_time = time.time()
elapsed_ms = (end_time - start_time) * 1000

print(f"GCD of {m} and {n}: {result}")
print(f"Elapsed Time: {elapsed_ms:.3f} ms")
```

```python
import time
def fibo_rec(n):
    if n < 2:
        return n
    else:
        return fibo_rec(n - 1) + fibo_rec(n - 2)
n=35
start_time = time.time()
result = fibo_rec(n)
end_time = time.time()
elapsed_ms = (end_time - start_time) * 1000
print(f"Fibonacci of {n}: {result}")
print(f"Elapsed Time: {elapsed_ms:.3f} ms")
```

```python
import time
def fibo_iter(n):
    i, j = 1, 0

    for k in range(n):
        j = i + j
        i = j - i
    return j
n=35
start_time = time.time()
result = fibo_rec(n)
end_time = time.time()
elapsed_ms = (end_time - start_time) * 1000
print(f"Fibonacci of {n}: {result}")
print(f"Elapsed Time: {elapsed_ms:.3f} ms")
```

```python
# write a function named as bubble_sort for Bubble-sort:
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap the elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Testing the bubble sort with your provided sample data
n = 100
raw_data = [int(random.random() * n) for _ in range(n)]
sorted_data = bubble_sort(raw_data.copy())
```
