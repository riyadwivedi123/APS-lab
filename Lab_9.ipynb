# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

2. Display shortest path with total distance.
3. Add two more nodes in graph.
4. Test algorithm with different source nodes.

```python
def dijkstra_with_path(graph, start):
    distances = {node: float('inf') for node in graph}
    predecessors = {node: None for node in graph}
    distances[start] = 0
    visited = set()

    while len(visited) < len(graph):
        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:
            break

        visited.add(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors

def get_shortest_path(predecessors, start_node, end_node):
    path = []
    current = end_node
    while current is not None and current in predecessors:
        path.insert(0, current)
        if current == start_node:
            break
        current = predecessors[current]

    if path and path[0] == start_node:
        return path
    else:
        return []

user_graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 3},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 7},
    'E': {'B': 3, 'D': 7, 'F': 6},
    'F': {'C': 8, 'E': 6}
}


print("--- Testing Dijkstra from source 'A' ---")
start_node_A = 'A'
distances_A, predecessors_A = dijkstra_with_path(user_graph, start_node_A)

for node in user_graph:
    path = get_shortest_path(predecessors_A, start_node_A, node)
    if path:
        print(f"Shortest distance from {start_node_A} to {node} = {distances_A[node]}. Path: {' -> '.join(path)}")
    else:
        print(f"Node {node} is not reachable from {start_node_A}.")

print("\n--- Testing Dijkstra from source 'B' ---")
start_node_B = 'B'
distances_B, predecessors_B = dijkstra_with_path(user_graph, start_node_B)

for node in user_graph:
    path = get_shortest_path(predecessors_B, start_node_B, node)
    if path:
        print(f"Shortest distance from {start_node_B} to {node} = {distances_B[node]}. Path: {' -> '.join(path)}")
    else:
        print(f"Node {node} is not reachable from {start_node_B}.")
```
